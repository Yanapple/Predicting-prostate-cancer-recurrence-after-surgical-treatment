Требуется: пациенты с рецидивом после операции (* простатэктомия)

[Операцию не рекомендуется проводить, если](https://medconsonline.com/ru/blog/rak-prostaty-priznaki-diagnostika):
- Имеется много разнолокализованных метастаз (вторичные очаги рака)
- При назначении паллиативной химиотерапии, когда рак не может быть удален уже из-за прорастания в разные органы массово (4 стадия заболевания)
- 7 < org_clinical_drug_prad(gleason_score) < 10
	Эти данные не дают гаранта отказа от операции. Наоборот, возможна радикальная простатэктомия.. за исключением балла 10.
- Прорастание рака в соседние органы, о чем свидетельствуют данные: 
	1) org_clinical_drug_prad(pathologic_T) >= T4*  и  org_clinical_drug_prad(pathologic_N) = 0  и  org_clinical_drug_prad(pathologic_M) = 0 
	ИЛИ  
	2) org_clinical_drug_prad(pathologic_N) = 1 и org_clinical_drug_prad(pathologic_M) = 0 
	ИЛИ 
	3) org_clinical_drug_prad(pathologic_M) = 1
То есть операцию проводить можно и рекомендуется при org_clinical_drug_prad(pathologic_T) < T4*, БЕЗ N (метастазы) и БЕЗ M (отдаленные метастазы)
Оценка ведется по патологии, а не клиническому заключению в датасете вследствие недостоверности клинических данных. Стараемся по возможности явно недостоверные данные исключить.

[PHI – индекс здоровья простаты](https://kdl.ru/analizy-i-tseny/indeks-zdorovya-prostati-prostate-health-index-phi-obshiy-prostatospetsificheskiy-antigen) – тест, который определяет уровень общего ПСА, процент свободного ПСА, - 2проПСА и рассчитывает показатели, с высокой точностью указывающие на риск развития рака. Это возможно благодаря тому, что учёные нашли и выделили форму – 2проПСА. Она вырабатывается злокачественными клетками агрессивной опухоли, и на ранней стадии может не влиять на уровень общего ПСА.
Рецидив наблюдается [при определенном уровне PCA\ПСА и PHI ](https://kdl.ru/patient/blog/psa-ili-phi#:~:text=%D0%A3%D1%80%D0%BE%D0%B2%D0%B5%D0%BD%D1%8C%20PHI%20%D0%BC%D0%B5%D0%BD%D0%B5%D0%B5%2025%20%D0%B3%D0%BE%D0%B2%D0%BE%D1%80%D0%B8%D1%82,%D0%BF%D1%80%D0%BE%D1%81%D1%82%D0%B0%D1%82%D1%8B%20%D0%B8%20%D1%82%D1%80%D0%B5%D0%B1%D1%83%D0%B5%D1%82%20%D0%BF%D1%80%D0%BE%D0%B2%D0%B5%D0%B4%D0%B5%D0%BD%D0%B8%D1%8F%20%D0%B1%D0%B8%D0%BE%D0%BF%D1%81%D0%B8%D0%B8.)в большинстве случаев:
1) Концентрацию ПСА выше 4 нг/мл обнаруживают примерно у 80-90% больных раком предстательной железы, а на уровне 2-10 нг\мл указывает на риск такового или наличие
2) При PHI > 25 риск наличия рака высокий
Для качественности полученного результата взятие образцов PCA рекомендуется 
- Через месяц после биопсии простаты и цистоскопии
- Через 10 дней после трансректального УЗИ (ТРУЗИ) и массажа предстательной железы
- За сутки воздержаться от полового акта
Рекомендуется обследование мужчинам старше 40 лет.
После удаления рака значение PCA -> 0 в течение 2-3 недель. [Резкое повышение после говорит о наличие рецидива.](https://wiki.nenaprasno.ru/a/chto-delat-esli-proizoshel-retsidiv-1)
